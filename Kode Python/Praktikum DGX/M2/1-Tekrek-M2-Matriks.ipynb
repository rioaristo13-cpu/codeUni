# Pertemuan II - Python Matriks dan NumPy Array

Matriks adalah struktur data dua dimensi dimana angka-angka disusun menjadi baris dan kolom. Contohnya:

<img src="https://cdn.programiz.com/sites/tutorial2program/files/matrix.jpg">

Matriks ini adalah matriks 3x4 (dibaca "tiga kali empat") karena memiliki 3 baris dan 4 kolom.

### Python Matriks
Python tidak memiliki tipe built-in untuk matriks. Namun, list di dalam list dapat digunakan untuk merepresentasikan matriks. Contohnya sebagai berikut:

```python
A = [[1, 4, 5],
    [-5, 8, 9]]
```

Kita dapat memperlakukan list ini sebagai matriks yang memiliki 2 baris dan 3 kolom.

<img src="https://cdn.programiz.com/sites/tutorial2program/files/matrix-python-example.jpg">

**Contoh Kode Program**

```python
A = [[1, 4, 5, 12],
    [-5, 8, 9, 0],
    [-6, 7, 11, 19]]
```

Untuk mencetak atau mendapatkan seluruh nilai yang ada pada matriks di atas dapat digunakan perintah:

```python
print("A =", A)
```

Selain itu, untuk mendapatkan elemen tertentu dari matriks tersebut dapat dilakukan dengan cara:

```python
# Mengakses seluruh elemen pada baris ke-2 (index dimulai dari 0)
print("A[1] =", A[1])

# Mengakses elemen ke-3 dari baris ke-2 (baris index 1, kolom index 2)
print("A[1][2] =", A[1][2])

# Mengakses elemen terakhir dari baris ke-1
# (-1 berarti index terakhir dalam Python)
print("A[0][-1] =", A[0][-1])
```

### NumPy Array
NumPy adalah library Python yang digunakan untuk komputasi ilmiah. Library ini menyediakan dukungan untuk objek array berdimensi-N (multidimensi).
<br><br>
Contoh:

```python
import numpy as np

a = [1, 2, 3])

print(a)

print(type(a))
```

Seperti yang Anda lihat, kelas array NumPy disebut `ndarray`.

<br>

### Bagaimana cara membuat array NumPy?
Ada beberapa cara untuk membuat array NumPy.

1. Array bilangan bulat, float dan bilangan kompleks.

```python
# Membuat array 2 dimensi (2 baris, 3 kolom) berisi bilangan bulat
A = [[1, 2, 3], [3, 4, 5]])
print(A)
```

```python
# Membuat array 2 dimensi (2 baris, 3 kolom) berisi bilangan float (desimal)
A = np.array([[1.1, 2, 3], [3, 4, 5]])
print(A)
```

```python
# Membuat array 2 dimensi (2 baris, 3 kolom) berisi bilangan kompleks
A = np.array([[1, 2, 3], [3, 4, 5]], dtype=complex)
print(A)
```

2. Array angka nol dan satu

```python
zeroes_array = np.zeros( (2, 3) )
print(zeroes_array)
```

Selain itu, kita juga dapat menetapkan `dtype` menjadi 32 bit (4 byte). Karena array ini dapat mengambil nilai dari <code>-2<sup>-31</sup></code> hingga <code>2<sup>-31</sup>-1</code>.

```python
ones_array = np.ones( (1, 5), dtype=np.int32 ) # menentukan dtype
print(ones_array)
```

### Operasi Matriks

Pada contoh program selanjutnya akan dibuat 3 jenis program yaitu:
1. Penjumlahan dua buah matriks,
2. Perkalian dua buah matriks dan
3. Transpos sebuah matriks.

Jika menggunakan program python biasa, kita memerlukan `nested lists` untuk membuat program tersebut. Namun ada cara yang lebih baik yaitu menggunakan library NumPy.

#### **Penambahan Dua Matriks**

Kita dapat menggunakan operator `+` untuk menambahkan elemen yang sesuai dari dua matriks NumPy.

```python
A = np.array(
    [[3, 4],
     [2, 1]])

B = np.array(
    [[1, 5],
     [3, 7]])

C = A + B

print(C)
```

#### **Perkalian Dua Matriks**

Untuk mengalikan dua matriks, kita menggunakan metode `dot()`.

**Catatan**: `*` digunakan untuk perkalian array (perkalian elemen yang sesuai dari dua array) bukan perkalian matriks.

```python
A = np.array(
    [[3, 4],
     [2, 1]])

B = np.array(
    [[1, 5],
     [3, 7]])

C = A.dot(B)

print(C)
```

---
#### **Transpose Matriks**

Kita dapat menggunakan `numpy.transpose` untuk menghitung transpose matriks.

```python
A = np.array(
    [[3, 4],
     [2, 1]])

print(A.transpose())
```

## **Tugas**

Pada contoh program diatas, telah dibuat tiga program menggunakan library NumPy, yaitu:
1. Penjumlahan dua matriks
2. Perkalian dua matriks
3. Transpose matriks

Tugas Anda adalah membuat kembali ketiga program tersebut (nilai bebas) **tanpa menggunakan library NumPy**, melainkan menggunakan struktur data dasar Python (seperti list) dan perulangan.

```python
# 1. Penjumlahan dua matriks
# CODE HERE ...
a =[[6,7,4],[1,2,3]]
b =[[5,8,2],[7,6,5]]

print ("A =", a)
print ("B =", b)

hasil = []
for i in range(len(a)):
    baris = []
    for j in range(len(a[i])):
        baris.append(a[i][j] + b[i][j])
    hasil.append(baris)
    
print ("\nHasil A + B =",hasil)
```

```python
# 2. Perkalian dua matriks
# CODE HERE ...
a = [[6,7,4],[1,2,3]]
b = [[5,8,2],[7,6,5]]

print ("A =", a)
print ("B =", b)

hasil = []
for i in range(len(a)):
    baris = []
    for j in range(len(a[i])):
        baris.append(a[i][j] * b[i][j])
        
    hasil.append(baris)
    
print ("\nHasil A * B =",hasil)
```

```python
# 3. Transpose matriks
# CODE HERE ...
a = [[6,7,4],[1,2,3]]
b = [[5,8,2],[7,6,5]]

print ("A =", a)
print ("B =", b)

def transpose(matrix):
    hasil = []
    for i in range(len(matrix[0])):
        baris = []
        for j in range(len(matrix)):
            baris.append(matrix[j][i])
        hasil.append(baris)
    return hasil

print (
f"""
Transpose A = {transpose(a)}
Transpose B = {transpose(b)}
""")

```
